```python
import matplotlib.pyplot as plt
from matplotlib import style

style.use("ggplot")#pehle to colur change kar dia(ggplot)......phir nightmode ho gya(dark_backgrouond)....backi dkeh lena

x=[1,2,3,4]
y=[7,6,8,9]

plt.plot(x,y,linewidth=2,color='c')
plt.title('my graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
#style.use("ggplot")

x1=[5,8,10]
y1=[12,16,6]

x2=[6,9,11]
y2=[6,15,7]

plt.plot(x1,y1,linewidth=2,color='c',label='first one')
plt.plot(x2,y2,linewidth=2,color='y',label='second one')

plt.title('my graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.grid(color='k')
plt.show()
```

```python
#bar

import matplotlib.pyplot as plt
plt.bar([1,3,5,7,9],[5,2,7,8,2])
plt.bar([2,4,6,8,10],[8,6,2,5,6])

plt.xlabel('my x')
plt.ylabel('my y')
plt.title('my 1stbar',color='k')

plt.show()
```

```python
#histogram

import matplotlib.pyplot as plt

px=[22,33,44,34,5,4,34,53,43,44,32,33,67,22,78,87,90,22,11,23,34,54,45,66,67,76,54,23,45,54,67,76,89,76,24]

py=[0,10,34,54,67,25,22,9,52,68,63,25,43,67,41,88,54,73,73,76,10,23]


plt.hist(py,rwidth=0.8,color='c')
plt.hist(px,rwidth=0.8,color='y')
plt.title('histogram')
plt.show()
```

```python
#scatterplot

x=[1,2,3,4,5,6,7,8]
y=[5,2,4,2,1,4,5,2]

plt.scatter(x,y,color='k',marker='x',s=70)
plt.title('scatter',color='k')
plt.show()
```

```python
#stack/area

days=[1,2,3,4,5]
sleeping=[7,8,8,11,9]
eating=[2,3,4,3,2]
working=[9,8,6,6,10]
playing=[6,5,6,4,3]
#for legend in stackplot
plt.plot([],[],color='m',label='sleeping',linewidth=5)
plt.plot([],[],color='c',label='eating',linewidth=5)
plt.plot([],[],color='y',label='working',linewidth=5)
plt.plot([],[],color='r',label='playing',linewidth=5)

plt.stackplot(days,sleeping,eating,working,playing,colors=['m','c','y','r'])
plt.title('Area Plot',color='k')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
```

```python
#string daal ke style bina style import
x=[1,2,3,4]
y=[1,4,9,16]

plt.plot(x,y,'k^:')#ro,g-,b,k^:....
plt.show()
```

```python
#pie chart

slices=[7,2,2,13]
activities=['sleeping','eating','working','playing']
cols=['c','m','r','b']
plt.pie(slices,labels=activities,colors=cols,shadow=True,autopct='%1.2f%%',startangle=90,explode=(0,0,0.4,0))
plt.title('piechart',color='k')
plt.show()
```

```python
import numpy as np

def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)

t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.02)

plt.subplot(211)
#subplot(nrows,ncols,plot_number)

#where nrows and n cols are used to notionally split the figure into nrows*ncols
#sub axes,and plt_number is used to identif y the paticular subplot that this function 
#is used to create within thenotional grid.plot_number starts at 1 , increments across 
#rows first and has a maximum of nrows *ncols.

plt.plot(t1,f(t1),'bo',t2,f(t2))
plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2))
plt.show()
```
